```python
import rasterio
from rasterio.plot import show
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Open the file:
with rasterio.open('media/aemet.tif') as src:
    transform = src.transform
    r = src.read(1)
    g = src.read(2)
    b = src.read(3)
    transform = src.transform
    crs = src.crs

    profile = src.profile

px_coords = (225,200)

lon, lat = transform * px_coords

plt.figure(figsize=(10, 10))
plt.imshow(np.dstack([r, g, b]))
plt.title('Horizontally Flipped RGB Image')
plt.scatter(px_coords[0], px_coords[1], c='red',s=100)  # Plot the pixel
plt.text(px_coords[0]+10, px_coords[1]-20, f'{px_coords[1]}, {lat,lon})', color='blue')
plt.legend()
plt.show()
```
