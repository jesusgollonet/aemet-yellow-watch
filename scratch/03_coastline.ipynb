```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from common_lib.util import rgb_to_hsv, color_contours
```

```python
img = cv.imread('../img/aemet-yellow.png', cv.IMREAD_COLOR)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB) )
```

```python

hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
brown = (255,227,173)
print(rgb_to_hsv(brown))
# Step 3: Define the color range for land (beige/brown)
lower_land = np.array([15, 60, 200])  # Adjusted lower bounds (narrower)
upper_land = np.array([25, 120, 255])
# Step 4: Create a mask using the color range
mask = cv.inRange(hsv, lower_land, upper_land)

plt.imshow(mask, cmap='gray')
# Step 5: Apply the mask to the original image
land_areas = cv.bitwise_and(img, img, mask=mask)
#plt.imshow(cv.cvtColor(land_areas, cv.COLOR_BGR2RGB) )

contours, _ = cv.findContours(mask, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

# Step 6: Draw the contours on the original image

cv.drawContours(img, contours, -1, (0, 255, 0), 2)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB) )
#plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB) )


```
