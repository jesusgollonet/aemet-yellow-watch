```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 5))
img = cv.imread('images/2024072400+108_aewam_alb_martot.png')
img2 = cv.imread('images/pre/2024072400+060_aewam_alb_martot.png')
ax[0].imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
ax[1].imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
```

```python
diff = cv.absdiff(img, img2)
plt.imshow(cv.cvtColor(diff, cv.COLOR_BGR2RGB))
```
